# Tensor

```python
import torch
import numpy as np
```

```python
a = [i for i in range(5)]
print(a)
print(type(a))
```

```python
# turn list to tensor
b = torch.tensor(a,dtype=torch.float32) # dtype=torch.float32 
print(b)
print(b.shape)
print(b.device)
print(b.dtype)
```

```python
# bumpy array
c = np.random.normal(0,2,size=(2,5))
c
```

```python

# turn array numpy to tensor
d = torch.from_numpy(c)
d
```

```python
# turn tensor to numpy
e = d.numpy()
e
```

```python
# add and sub and div
d = torch.rand(1,5)
print(d)

add = torch.add(d, 20)
print(add)

sub = torch.sub(d, 0.8)
print(sub)

div = torch.div(d, 2)
print(div)
```

```python
# slicing
d = torch.rand(3,5)
print(d)
print(d[:,2])
print(d[1,1])
```

```python
a = torch.tensor([[1, 2],[3, 4]])
print(a)
print(a.shape)
```
